# Importing Libraries

```python
#import all the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Importing the Dataset

```python
#import the dataset

dataset = pd.read_csv('C:\\Users\\Chandu\\Desktop\\Datasets\\SAT-GPA.csv')
```

# Observing the Dataset

```python
#observe the dataset

dataset
```

This dataset gives us the information about the SAT scores and GPA.
By using the simple linear regression, let's predict the SAT score based on the GPA.

# Preprocessing the Dataset

```python
#Getting the basic information of the dataset

dataset.info()
```

This dataset has 84 entries of SAT and GPA values of which SAT is of integer and GPA is of float datatype.

```python
#Let's seperately check for only null values

dataset.isnull().sum()
```

This tells us that none of the columns in the dataset has null values.

# Visualizing the Data

```python
#Lets draw a graph between SAT and GPA scores

plt.figure(figsize=(40,20))
plt.plot(dataset["GPA"], dataset["SAT"])
plt.xlabel('GPA')
plt.ylabel('SAT')
plt.show()
```

# Dividing into input and output

Since, the students having higher GPA will probably have a higher SAT score, let's take the GPA as the independent factor(x) and the SAT score as the dependent factor(y)

```python
#Let's take the GPA as 'x-values'(input) and SAT scores as 'y-values'(output).

x = dataset["GPA"].values
y = dataset["SAT"].values
```

```python
#Let's see the shape of x

x.shape
```

Since our x(input) values should be a 2-D array, let us reshape this 1-D array into a 2-D array.

```python
#Let's reshape x

x = x.reshape(84,1)
x.shape
```

```python
#Let's see the x values
x
```

```python
#Let's see the y values

y
```

```python
#Let's see the shape of y

y.shape
```

# Dividing the dataset into Train and Test parts

```python
#Let's split the dataset into testing and training part

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
#Let's see the size of x_train and y_train dataparts

x_train.size, y_train.size
```

```python
#Let's see the shape of x_train and y_train dataparts

x_train.shape, y_train.shape
```

```python
#Let's see the size of x_test and y_test dataparts

x_test.size, y_test.size
```

# Running a Regressor

```python
#Let's run a regressor to get the predicted values of GPA based on the SAT scores.

lr = LinearRegression()
```

# Fitting the model

```python
#Let's fit the model by mapping the input with output of the training dataset.

lr = lr.fit(x_train, y_train)
```

# Predicting the output

```python
#Let's predict the output values(y_test) based on the input values(x_test)

y_pred = lr.predict(x_test)
y_pred
```

```python
#Let's see the actual output values(y_test)

y_test
```

```python
#Let's see the difference between the actual and the predicted output values

y_test - y_pred
```

```python
#Let's see the graphical representation of the actual and the predicted output values

#let's point the actual output values(y_test) vs the actual input values(x_test)
plt.scatter(x_test, y_test)

#let's point the predicted output values(y_pred) vs the actual input values(x_test)
plt.plot(x_test, y_pred, 'r')
```

# Evaluating the Accuracy score of the model

```python
#Let's check the accuracy score of the model

r2_score(y_test, y_pred)
```
